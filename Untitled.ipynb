```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F
```

```python
# 数据预处理
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
```

```python
# 加载数据集
os.environ["CUDA_VISIBLE_DEVICES"]="0"
dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
```

```python
# 提取数据和标签
data = [dataset[i][0].numpy() for i in range(len(dataset))]
labels = [dataset[i][1] for i in range(len(dataset))]
```

```python
# 使用train_test_split进行划分
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
# 转换为Tensor
x_train = torch.tensor(x_train)
x_test = torch.tensor(x_test)
y_train = torch.tensor(y_train)
y_test = torch.tensor(y_test)
```

```python
print(f'x_train shape: {x_train.shape}')
print(f'x_test shape: {x_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
# 归一化
x_train = (x_train - x_train.min()) / (x_train.max() - x_train.min())
x_test = (x_test - x_test.min()) / (x_test.max() - x_test.min())
```

```python
# 显示归一化后的图片
def imshow(img, title=None):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    if title:
        plt.title(title)
    plt.show()
```

```python
# 从训练集中取一张图片进行显示
image = dataset[0][0]  # 原始图片
normalized_image = (image - image.min()) / (image.max() - image.min())  # 归一化后的图片

# 显示原始图片
imshow(image, 'Original Image')

# 显示归一化后的图片
imshow(normalized_image, 'Normalized Image')
```

```python
# 指定使用 GPU 1
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
```

```python
# 检查是否有 GPU 可用
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# 定义CNN模型
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 8 * 8, 512)
        self.fc2 = nn.Linear(512, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 8 * 8)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

net = SimpleCNN().to(device)
```

```python
# 定义损失函数和优化器
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# 训练模型
for epoch in range(10):  # 迭代次数
    running_loss = 0.0
    for i in range(len(x_train)):
        inputs = x_train[i].unsqueeze(0)
        labels = y_train[i].unsqueeze(0)
        optimizer.zero_grad()
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 200 == 199:  # 每200个小批量打印一次
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] loss: {running_loss / 200:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
# 保存模型
torch.save(net.state_dict(), 'simple_cnn.pth')
```

```python
net.load_state_dict(torch.load('simple_cnn.pth'))
net.eval()  # 设置模型为评估模式
```

```python
# 测试模型
correct = 0
total = 0
with torch.no_grad():
    for i in range(len(x_test)):
        images = x_test[i].unsqueeze(0)
        labels = y_test[i].unsqueeze(0)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the 10000 test images: {100 * correct / total} %')
```

```python
# 定义RNN模型
class SimpleRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(SimpleRNN, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out

input_size = 32 * 32 * 3  # CIFAR-10 图片大小为32x32，3个通道
hidden_size = 128
output_size = 10  # CIFAR-10 有10个类别
num_layers = 2

net = SimpleRNN(input_size, hidden_size, output_size, num_layers).to(device)
```

```python
# 定义损失函数和优化器
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# 训练模型
for epoch in range(10):  # 迭代次数
    running_loss = 0.0
    for i in range(len(x_train)):
        inputs = x_train[i].view(1, -1, input_size).to(device)  # 调整输入形状
        labels = y_train[i].unsqueeze(0).to(device)
        optimizer.zero_grad()
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 200 == 199:  # 每200个小批量打印一次
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] loss: {running_loss / 200:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
# 保存模型
torch.save(net.state_dict(), 'simple_rnn.pth')
```

```python
net.load_state_dict(torch.load('simple_rnn.pth'))
net.eval()  # 设置模型为评估模式
```

```python
# 测试模型
correct = 0
total = 0
with torch.no_grad():
    for i in range(len(x_test)):
        images = x_test[i].view(1, -1, input_size)  # 调整输入形状
        labels = y_test[i].unsqueeze(0)
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Accuracy of the network on the 10000 test images: {100 * correct / total} %')
```
